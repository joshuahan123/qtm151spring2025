```python
# Write your answer
china_emissions = pd.read_csv('data/china_co2_emissions.csv')
china_year = china_emissions["year"]
china_total_emissions = china_emissions["total_emissions"]
plt.scatter(x=china_year, y=china_total_emissions)
plt.title("Total Emissions per Year in China")
plt.xlabel("Year")
plt.ylabel("Total Emissions")
plt.show()


germany_emissions = pd.read_csv('data/germany_co2_emissions.csv')
germany_year = germany_emissions["year"]
germany_total_emissions = germany_emissions["total_emissions"]
plt.scatter(x=germany_year, y=germany_total_emissions)
plt.title("Total Emissions per Year in Germany")
plt.xlabel("Year")
plt.ylabel("Total Emissions")
plt.show()
```
